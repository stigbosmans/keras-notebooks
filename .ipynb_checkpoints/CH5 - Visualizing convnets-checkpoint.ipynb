# Visualize Convnet Filters

```python
from keras.applications import VGG16
from keras import backend as K
model = VGG16(weights='imagenet', include_top=False)
model.summary()
```

We'll now get output from a certain layer. We set this output as loss, because if we maximize this loss, we can assume that the formed pattern in this layer will be directly related to the maximal activation values of this layer. 

After obtaining the first output we calculate the gradients based on the input image. We add these gradients (instead of subtracting like we do with standard backpropagation gradient descent) to the input image and this should then increase the loss. 
After 40 times we should obtain maximal loss, and we can visualize the pattern

```python
def deprocess_image(x):
    x -= x.mean()
    x /= (x.std()) + 1e-5
    x *= 0.1
    x += 0.5
    x = np.clip(x, 0, 1)
    x *= 255
    x = np.clip(x, 0, 255).astype('uint8')
    return x

def generate_pattern(layer_name, filter_index, size=150):
    #Load output of a specific layer and set as loss. 
    layer_output = model.get_layer(layer_name).output
    loss = K.mean(layer_output[:, :, :, filter_index])
    
    #find gradients based on a given input and apply L2 regularization
    grads = K.gradients(loss, model.input)[0]
    grads /= (K.sqrt(K.mean(K.square(grads))) + 1e-5)
    
    #Tensorflow lambda calculates loss and grads based on given input
    iterate = K.function([model.input], [loss, grads])
    
    #Start from noisy grayscale image
    input_img_data = np.random.random((1, size, size, 3)) * 20 + 128. 
    print(iterate)
    step = 1.
    for i in range(40):
        loss_value, grads_value = iterate([input_img_data])
        input_img_data += grads_value * step
    
    img = input_img_data[0]
    return deprocess_image(img)
```

Note that all the K.something methods are evaluated as placeholders. Only when we run iterate function (K.function) 
will we run the tensorflow calculations. The output of the function will be a numpy array with real values.

deprocess_image normalizes the values to the same range by subtracting mean and std and then clip it between 0 and 255

```python
plt.imshow(generate_pattern('block4_conv1', 0))
```

Above image shows the pattern that the filter 0 of block 3, conv1 responds to maximally

# GRAD-CAM

Visualize which parts of an image contribute the most to the activation of a given class

```python
from keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input, decode_predictions
import numpy as np
from matplotlib.pyplot import plt
```

```python
from keras.applications.vgg16 import VGG16
model = VGG16(weights='imagenet')
```

```python
model.summary()
```

```python
img = image.load_img('elephant.jpg', target_size=(224, 224))
```

```python
plt.imshow(img)
```

```python
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)
```

```python
preds = model.predict(x)
decode_predictions(preds)
```

```python
# Find output index of elephant
np.argmax(preds)
```

```python
elephant_output = model.output[:, 386]
```

```python
elephant_output
```

```python
last_conv_layer = model.get_layer('block5_conv3')
```

```python
grads = K.gradients(elephant_output, last_conv_layer.output)[0] #The higher the gradient in a given dimension, the higher it activates
pooled_grads = K.mean(grads, axis = (0, 1, 2)) #grads default shape is 1,14,14,512, we calculate the mean of all grads of each individual channel
iterate = K.function([model.input], [pooled_grads, last_conv_layer.output[0]])
```

```python
pooled_grads_value, conv_layer_output_value = iterate([x])
```

```python
pooled_grads_value.shape
```

```python
conv_layer_output_value.shape
```

Now we multiply each channel of the output with the gradients or by how important it is in regards to elephant case.
Large grads means large influence on prediction. 

```python
for i in range(512):
    conv_layer_output_value[:, :, i] *= pooled_grads_value[i] #multiply average grad value (=importance) with all pixels in a certain channel
```

```python
heatmap = np.mean(conv_layer_output_value, axis=-1) # take average of all 512 channels
```

```python
heatmap.shape
```

```python
heatmap = np.maximum(heatmap, 0) #remove all values <0
heatmap /= np.max(heatmap) # scale between 0 and 1
```

```python
plt.matshow(heatmap)
```

```python
plt.imshow(img)
```

Now let's add the activations on top of the original image

```python
import cv2
```

```python
img = cv2.imread('elephant.jpg')
heatmap = cv2.resize(heatmap, (img.shape[1], img.shape[0]))
heatmap = np.uint8(255 * heatmap) #converts to white scale
heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)
superimposed = heatmap * 0.4 + img
cv2.imwrite('superimpose_elephant.jpg', superimposed)
```
